# **ЦІКАВІ ТА МАГІЧНІ КОДИ НА PYTHON**

Код Магічний Генератор Паролів:

```python
import random
import string

password_length = 12
password = ''.join(random.choices(string.ascii_letters + string.digits + string.punctuation, k=password_length))

print("Your magical password:", password)
```

Цей код створює випадковий пароль довжиною 12 символів, використовуючи букви (великі та малі), цифри та символи.

---



Код Магічний Генератор Віршів:

```python
import random

beginning = ["Amidst the bright summer,", "In the clear night sky,", "In the quiet and misty valley,"]
middle = ["silver stars are playing.", "the moon shines like a fiery ball.", "the trees whisper their secrets,"]
end = ["A bright star shines and the night is silent.", "Dreams fly to a distant world.", "Here the magical world ends."]

random_poem = random.choice(beginning) + " " + random.choice(middle) + " " + random.choice(end)
print("Your magical poem:")
print(random_poem)
```

Цей код генерує випадкові вірші, обираючи випадкові фрази з кожної частини вірша.

---



Код Магічний Калькулятор:

```python
def magical_calculator(operation, number1, number2):
    if operation == "+":
        return number1 + number2
    elif operation == "-":
        return number1 - number2
    elif operation == "*":
        return number1 * number2
    elif operation == "/":
        if number2 != 0:
            return number1 / number2
        else:
            return "Division by zero is not possible!"

operation = input("Choose operation (+, -, *, /): ")
number1 = float(input("Enter the first number: "))
number2 = float(input("Enter the second number: "))

result = magical_calculator(operation, number1, number2)
print("Result:", result)
```

Цей код створює простий калькулятор, який може виконувати операції додавання, віднімання, множення та ділення.

**ОСОБЛИВИЙ КОД ЯКИЙ ПЕРЕСУВАЄ ТЕКСТ!!**

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import time

output_text = "This is moving text!"
text_widget = widgets.HTML(value=output_text)
left_button = widgets.Button(description='← Left')
right_button = widgets.Button(description='Right →')

def move_left(_):
    shifted_text = text_widget.value[-1] + text_widget.value[:-1]
    text_widget.value = shifted_text

def move_right(_):
    shifted_text = text_widget.value[1:] + text_widget.value[0]
    text_widget.value = shifted_text

left_button.on_click(move_left)
right_button.on_click(move_right)

display(widgets.HBox([left_button, right_button]))
display(text_widget)
```

Цей код створює віджети кнопок "← Вліво" і "Вправо →", які дозволяють рухати
текст відповідно вліво та вправо. По натисканні на відповідні кнопки, текст зсувається у відповідному напрямку.
